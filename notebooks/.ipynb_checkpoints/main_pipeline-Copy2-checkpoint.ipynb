# 数据处理

# 2. 超参数设置

```python
# ====== Config ======
PATCH_SIZE = (64, 64, 64)
BATCH_SIZE = 2
EPOCHS = 10
LR = 1e-4
DATA_DIR = "/root/lanyun-tmp/amos_dataset/amos22/npy_patches"
CHECKPOINT_PATH = "checkpoints/anatomask_real.pth"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

#  3. 模型构建

```python
cnn = Simple3DCNNEncoder(in_channels=1, base_channels=64)
encoder = SparseEncoder(cnn, input_size=PATCH_SIZE, sbn=False)
decoder = LightDecoder(up_sample_ratio=8, width=256, sbn=False)
model = SparK(encoder, decoder, mask_ratio=0.6).to(DEVICE)

optimizer = optim.Adam(model.parameters(), lr=LR)
```

# 4. 数据加载

```python
dataset = Real3DMedicalDataset(DATA_DIR)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
```

# 5. 启动训练

```python
train(model, optimizer, DEVICE, epochs=EPOCHS, dataloader=loader, save_path=CHECKPOINT_PATH)
```

# 6. 加载训练好的模型（如已训练过）

```python
# 如果你想加载已有的模型，设置这个变量为 True
LOAD_PRETRAINED = True

if LOAD_PRETRAINED:
    model, optimizer, start_epoch = load_checkpoint(model, optimizer, CHECKPOINT_PATH, DEVICE)
    print(f"✅ 成功加载模型！从 epoch {start_epoch+1} 继续")
else:
    print("⚠️ 未加载任何 checkpoint，重新开始训练")
```

```python
import matplotlib.pyplot as plt
import torch

def visualize_segmentation_prediction(model, dataset, device, slice_idx=None, axis=2):
    model.eval()
    with torch.no_grad():
        x, y = dataset[0]  # 一张 patch
        x = x.unsqueeze(0).to(device)  # [1, 1, 64, 64, 64]
        y = y.cpu().numpy()           # [8, 8, 8]

        pred = model(x)
        pred_label = torch.argmax(pred, dim=1).squeeze(0).cpu().numpy()  # [8, 8, 8]

        # 选择 slice
        if slice_idx is None:
            slice_idx = y.shape[axis] // 2

        def get_slice(vol):
            return {
                0: vol[slice_idx, :, :],
                1: vol[:, slice_idx, :],
                2: vol[:, :, slice_idx],
            }[axis]

        image_slice = x.squeeze().cpu().numpy()[32]  # 原图切一层
        label_slice = get_slice(y)
        pred_slice = get_slice(pred_label)
        diff_slice = (label_slice != pred_slice).astype(int)

        # 可视化
        titles = ["Image", "Ground Truth", "Prediction", "Error"]
        data = [image_slice, label_slice, pred_slice, diff_slice]

        plt.figure(figsize=(16, 4))
        for i in range(4):
            plt.subplot(1, 4, i+1)
            plt.imshow(data[i], cmap="gray" if i==0 else "viridis")
            plt.title(titles[i])
            plt.axis("off")
        plt.tight_layout()
        plt.show()
```

```python
visualize_segmentation_prediction(seg_model, seg_dataset, DEVICE, axis=2)
```
